# Testing a set of random numbers if they are uniformly distributed

```python
from random import *
from statistics import *

N = 10_000
data = [random() for i in range(N)]

corr = 0
for i in range(N-1):
    corr += data[i] * data[i+1]
corr = corr/N

print("Mean =", round(mean(data), 2))
print("Variance =", round(variance(data), 2))
print("Autocorrelation =", round(corr, 2))
```

# Generating the maximum-legth random sequence from the fou-bit LFSR

```python
seed = 0b_0001
num = seed

# Define masks
mask1 = 0b_0001
mask2 = 0b_0110
mask3 = 0b_0001

# Counter
period = 0

while True:
    print(num)
    
    temp1 = ((num >> 1) ^ (num << 0)) & mask1
    temp2 = ( num >> 1) & mask2
    temp3 = (num & mask3) << 3
    num = temp1 | temp2 | temp3
    
    period += 1
    
    if num == seed:
        break
        
print("Period =", period)
```

# Generating the Maximum-length random sequence from an eight-bit LFSR

```python
seed = 0b_00111000
num = seed

# Define masks
mask1 = 0b_00000010
mask2 = 0b_00000100
mask3 = 0b_00001000
mask4 = 0b_01110001
mask5 = 0b_00000001

# Counter
period = 0

while True:
    
    
    temp1 = ((num >> 1) ^ (num << 1)) & mask1
    temp2 = ((num >> 1) ^ (num << 2)) & mask2
    temp3 = ((num >> 1) ^ (num << 3)) & mask3
    temp4 = (num >> 1) & mask4
    temp5 = (num & mask5) << 7
    num = temp1 | temp2 | temp3 | temp4 | temp5
    
    period += 1
    print(num)
    if num == seed:
        break
        
print("Period =", period)
```

# Python program for generating a 3D scattered plot for the spectral test

```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = 65539
M = math.pow(2, 31)
seed = 123456

X = []
Y = []
Z = []

for i in range(1000):
    num1 = math.fmod(a * seed, M)
    num2 = math.fmod(a * num1, M)
    num3 = math.fmod(a * num2, M)
    seed = num2
    X.append(num1)
    Y.append(num2)
    Z.append(num3)
    
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X, Y, Z, c="b", marker="o")

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

# Python procedure for generating a lag plot for a random sequence

```python
import random as rnd
import pandas
from pandas.plotting import lag_plot
import matplotlib.pyplot as plt

s = pandas.Series([rnd.random() for i in range(10000)])

plt.figure()
lag_plot(s, c="grey", marker="o")
plt.xlabel("Random Number - s[i]")
plt.ylabel("Lag1(Random Number) - s[i+1]")
plt.show()
```
